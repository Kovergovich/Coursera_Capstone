## Exploring

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhoods', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

## Analyzing

```python
frunze_onehot = pd.get_dummies(frunze_venues[['Venue Category']], prefix="", prefix_sep="")
frunze_onehot['Neighborhoods'] = frunze_venues['Neighborhoods'] 
fixed_columns = [frunze_onehot.columns[-1]] + list(frunze_onehot.columns[:-1])
frunze_onehot = frunze_onehot[fixed_columns]
frunze_onehot.head()
```

```python
frunze_onehot.shape
```

```python
frunze_grouped = frunze_onehot.groupby('Neighborhoods').mean().reset_index()
frunze_grouped
```

```python
frunze_grouped.shape
```

```python
num_top_venues = 5

for hood in frunze_grouped['Neighborhoods']:
    print("----"+hood+"----")
    temp = frunze_grouped[frunze_grouped['Neighborhoods'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']
columns = ['Neighborhoods']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhoods'] = frunze_grouped['Neighborhoods']

for ind in np.arange(frunze_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(frunze_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

## Clustering

```python
kclusters = 5
frunze_grouped_clustering = frunze_grouped.drop('Neighborhoods', 1)
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(frunze_grouped_clustering)
kmeans.labels_[0:10] 
```

```python
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
frunze_merged = frunze_data
frunze_merged = frunze_merged.join(neighborhoods_venues_sorted.set_index('Neighborhoods'), on='Neighborhoods')
frunze_merged.head()
```

```python
missing_data = frunze_merged.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("") 
```

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]
markers_colors = []
for lat, lon, poi, cluster in zip(frunze_merged['Latitude'], frunze_merged['Longitude'], frunze_merged['Neighborhoods'], frunze_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[int(cluster)-1],
        fill=True,
        fill_color=rainbow[int(cluster)-1],
        fill_opacity=0.7).add_to(map_clusters)
map_clusters
```
